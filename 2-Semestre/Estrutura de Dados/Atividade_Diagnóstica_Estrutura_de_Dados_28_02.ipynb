1. Crie um projeto Python e Faça um projeto que leia a idade do usuário e classifique-o em:
- Criança – 0 a 10 anos
- Pré-Adolescente – 11 a 13 anos
- Adolescente – 14 a 17 anos
- Adulto – acima de 18 anos
Obs.: Se o usuário digitar um número negativo, ou maior que 150, mostre uma mensagem
informando que a idade é inválida

```python
def classificar_idade(idade):
    if 0 <= idade <= 10:
        return "Criança"
    elif 11 <= idade <= 13:
        return "Pré-Adolescente"
    elif 14 <= idade <= 17:
        return "Adolescente"
    elif idade >= 18:
        return "Adulto"
    else:
        return "Idade inválida"

def main():
    try:
        idade = int(input("Digite a sua idade: "))
        if 0 <= idade <= 150:
            classificacao = classificar_idade(idade)
            print(f"Você é classificado como: {classificacao}")
        else:
            print("Idade inválida. Por favor, insira uma idade entre 0 e 150 anos.")
    except ValueError:
        print("Por favor, insira um valor numérico válido para a idade.")

if __name__ == "__main__":
    main()
```

2. Em um novo projeto, efetue o cálculo da quantidade total de litros de combustível que um veículo usará em uma viagem.
Para obter o cálculo, o usuário deve fornecer os quilômetros total da viagem e a média de consumo do seu veículo.


```python
def calcular_litros_viagem(km_total, media_consumo):
    litros_necessarios = km_total / media_consumo
    return litros_necessarios

def main():
    try:
        km_total = float(input("Informe a quantidade total de quilômetros da viagem: "))
        media_consumo = float(input("Informe a média de consumo do veículo em km/l: "))

        if km_total > 0 and media_consumo > 0:
            litros_necessarios = calcular_litros_viagem(km_total, media_consumo)
            print(f"A quantidade total de litros de combustível necessários é: {litros_necessarios:.2f} litros")
        else:
            print("Os valores informados devem ser maiores que zero.")
    except ValueError:
        print("Por favor, insira valores numéricos válidos.")

if __name__ == "__main__":
    main()
```

3. Em um novo projeto, calcule a média de um aluno que cursou a disciplina de Estrutura de Dados.
Leia as notas N1, N2, N3, N4, N5 e N6, e com os valores lidos, calcule a média aritmética.
Depois, informe se o aluno foi aprovado, reprovado ou se terá que fazer uma prova substitutiva.
- Se a média estiver abaixo 4.5 o aluno está reprovado
- Se a média estiver entre 4.5 até 6, o aluno terá que fazer uma sub
- Se a média for maior do que 6, o aluno está aprovado


```python
def calcular_media(notas):
    return sum(notas) / len(notas)

def verificar_aprovacao(media):
    if media < 4.5:
        return "Reprovado"
    elif 4.5 <= media < 6:
        return "Prova Substitutiva"
    else:
        return "Aprovado"

def main():
    try:
        notas = [float(input(f"Informe a nota N{i + 1}: ")) for i in range(6)]

        if all(0 <= nota <= 10 for nota in notas):
            media = calcular_media(notas)
            situacao = verificar_aprovacao(media)
            print(f"A média do aluno é: {media:.2f}")
            print(f"Situação do aluno: {situacao}")
        else:
            print("As notas devem estar no intervalo de 0 a 10.")
    except ValueError:
        print("Por favor, insira valores numéricos válidos.")

if __name__ == "__main__":
    main()
```

4. Faça outro projeto que use uma estrutura de repetição que imprima a tabuada do número que a pessoa digitar. A tabuada será até o valor 10.
No início, peça ao usuário para ele digitar o número desejado, e com esse número você calcula a tabuada.
Ou seja, se a pessoa digitar 50, será feito:
50 x 1 = 50
50 x 2 = 100
50 x 3 = 150
...
50 x 10 = 500

```python
def calcular_tabuada(numero):
    for i in range(1, 11):
        resultado = numero * i
        print(f"{numero} x {i} = {resultado}")

def main():
    try:
        numero = int(input("Digite um número para calcular a tabuada: "))
        calcular_tabuada(numero)
    except ValueError:
        print("Por favor, insira um número inteiro válido.")

if __name__ == "__main__":
    main()
```

5. Em um novo projeto, crie uma aplicação para ser usado pelo Secretaria da Fazenda da prefeitura. O sistema deverá receber dois valores, sendo o tamanho de um imóvel, calcular a sua quantidade de metros quadrados e definir o valor do IPTU a ser pago pelo proprietário.
Considere o valor de IPTU de 1 m² sendo R$ 8,14.


```python
def calcular_iptu(tamanho_imovel):
    valor_por_metro_quadrado = 8.14
    iptu = tamanho_imovel * valor_por_metro_quadrado
    return iptu

def main():
    try:
        comprimento = float(input("Digite o comprimento do imóvel em metros: "))
        largura = float(input("Digite a largura do imóvel em metros: "))

        if comprimento > 0 and largura > 0:
            area_imovel = comprimento * largura
            valor_iptu = calcular_iptu(area_imovel)

            print(f"\nÁrea do imóvel: {area_imovel:.2f} m²")
            print(f"O valor do IPTU a ser pago é: R$ {valor_iptu:.2f}")
        else:
            print("As dimensões do imóvel devem ser maiores que zero.")
    except ValueError:
        print("Por favor, insira valores numéricos válidos para as dimensões do imóvel.")

if __name__ == "__main__":
    main()
```

6. Faça uma calculadora de IMC (Índice de Massa Corporal) e dê o resultado.


```python
def calcular_imc(peso, altura):
    imc = peso / (altura ** 2)
    return imc

def classificar_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    elif 30 <= imc < 35:
        return "Obesidade Grau I"
    elif 35 <= imc < 40:
        return "Obesidade Grau II"
    else:
        return "Obesidade Grau III"

def main():
    try:
        peso = float(input("Digite o peso em quilogramas: "))
        altura = float(input("Digite a altura em metros: "))

        if peso > 0 and altura > 0:
            imc = calcular_imc(peso, altura)
            situacao_imc = classificar_imc(imc)

            print(f"\nSeu IMC é: {imc:.2f}")
            print(f"Sua situação é: {situacao_imc}")
        else:
            print("Peso e altura devem ser valores positivos.")
    except ValueError:
        print("Por favor, insira valores numéricos válidos para peso e altura.")

if __name__ == "__main__":
    main()
```

7. Faça um programa para um Caixa Eletrônico. O programa deverá perguntar ao usuário o valor que deseja sacar e depois informar quantas notas de cada valor serão fornecidas.
As notas disponíveis serão as de 1, 5, 10, 100 e 200 reais.
O valor mínimo é de 10 reais e o máximo de 1500 reais.
O programa não deve se preocupar com a quantidade de notas existentes na máquina.
Exemplo 1: Para sacar a quantia de 456 reais, o programa fornece duas notas de 200, cinco notas de 50, uma nota de 5 e uma nota de 1;
Exemplo 2: Para sacar a quantia de 399 reais, o programa fornece uma nota de 200 reais, uma nota de 100, uma nota de 50, quatro notas de 10, uma nota de 5 e quatro notas de 1.

```python
def caixa_eletronico(valor_saque):
    notas_disponiveis = [200, 100, 50, 10, 5, 1]
    notas_contagem = {}

    if 10 <= valor_saque <= 1500:
        for nota in notas_disponiveis:
            quantidade_notas = valor_saque // nota
            if quantidade_notas > 0:
                notas_contagem[nota] = quantidade_notas
                valor_saque %= nota

        return notas_contagem
    else:
        return None

def main():
    try:
        valor_saque = int(input("Digite o valor que deseja sacar (entre 10 e 1500 reais): "))
        resultado = caixa_eletronico(valor_saque)

        if resultado is not None:
            print("\nNotas fornecidas:")
            for nota, quantidade in resultado.items():
                print(f"{quantidade} nota(s) de {nota} reais")
        else:
            print("Valor inválido. O valor do saque deve estar entre 10 e 1500 reais.")
    except ValueError:
        print("Por favor, insira um valor numérico válido para o saque.")

if __name__ == "__main__":
    main()
```
