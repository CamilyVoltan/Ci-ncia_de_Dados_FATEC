```python
from matplotlib import pyplot as plt
```

plt.***title*** = para adicionar título

plt.***label*** = etiqueta para a linha/coluna

plt.***tick_params*** = personalizar os parâmetros dos ticks (marcadores) em um gráfico. Essa função permite que você ajuste vários aspectos relacionados aos ticks, como sua aparência, posição, tamanho, cor, entre outros

plt.***grid*** = se refere a uma grade ou matriz bidimensional

plt.***show*** = mostra o gráfico, no colab não precisa, mas em outras IDEs é necessário

***fontsize*** = tamanho da fonte (letras)

plt.***xticks*** = refere-se aos marcadores ou rótulos ao longo do eixo x (horizontal) de um gráfico. Esses marcadores ajudam a identificar e entender a escala do eixo x

```python
y = [1,4,9,16,25]
x = [1,2,3,4,5]

plt.plot(x, y, linewidth=5, c='red')
plt.title('NÚMEROS AO QUADRADO',fontsize=20)
plt.xlabel('NÚMEROS',fontsize=10)
plt.ylabel('QUADRADO',fontsize=10)
plt.tick_params(axis='both',labelsize=14)
plt.grid(True)
plt.show(True)
```

```python
x = [1,2,3,4,5,6]
y = [2,1,6,4,8,5]
values = ['A','B','C','D','E','F']

#plt.style.use('bmh')
plt.plot(x, y, marker='o')
plt.xticks(x,values)
plt.show(True)
```

```python
x = ['JAN', 'FEV', 'MAR', 'ABR', 'MAIO', 'JUN']
y1 = [40,42,35,38.5,28,22]
y2 = [38,35,28,23,24,30]

#plt.style.use('seaborn')
plt.title('MÉDIA DE TEMPERATURAS')
plt.xlabel('MESES')
plt.ylabel('GRAUS CELSIUS')
  #plt.axis(ymin=0, ymax=50)
plt.plot(x, y1, label='2020') #quando coloca 'label' no plot x ou y, ele vai aparecer como uma legenda da plotagem, ex: ano 2020 e 2021
plt.plot(x, y2, label='2021')
plt.legend(fontsize=16,frameon=True,framealpha=0.2,facecolor='black')
  #frameon = controlar a exibição da moldura ao redor de um gráfico ou figura
  #framealpha = controla a transparência da tabela


plt.show(True)
```

```python
x = ['MAR', 'ABR', 'MAIO', 'JUN','JUL','AGO','SET']
y = [80, 40, 80, 110, 180, 120, 240]

plt.title('CONSUMO DE ENERGIA (W)')
plt.xlabel('WATTS')
plt.ylabel('MESES')
plt.plot(x, y, marker='o')
plt.grid(True)
plt.show(True)
```
