**USANDO RESHAPE**

**Sintaxe:** array.reshape(lin, col)

```python
import numpy as np
vetor = np.arange(10) #vetor de 10 elementos #arange = para criar uma sequência númerica
print(f'Exibindo o vetor e {vetor.size} elementos: {vetor}')
print()

#Criando a matriz
#a quantidade de linhas x colunas deve ser igual a quantidade de elementos
matriz = vetor.reshape(5, 2)

print(f'Exiindo a matriz de {matriz.size} elemetntos: \n {matriz}')
print(f'Exibindo a forma da matriz: {matriz.shape}')
```

```python
# reshape(numero de matriz, linhas, colunas)
# num_matrizes *linhas* colunas = quantidade de elementos
v = np.array(range(50)).reshape(2,5,5)

print('Shape = ', v.shape)
print('Número de dimensões = ', v.ndim) # v.ndim =
print('Número de elementos = ', v.dtype)
print('Matriz v = \n', v)
```

```python
# Multiplicação matricial entre matrizes

import numpy as np

A = np.array([[1, -2], [2, 4]])
B = np.array([[1, 2],[2, 4]])

print(f'Matriz A: {A}')
print(f'Matriz B: {B}')

print(f'\nMultiplicação AXB: {np.matmult(A, B)}')
print(f'\nMultiplicação')
```

```python
# Determinante de uma matriz e matriz inversa
import numpy as np

B = np.array([[2, 1],[5, 3]])

print(f'matriz B: \n {B}')
print(f'Determinante: {np.linalg.det(B):.2f}')
print(f'Inversa:\n {np.linalg.inv(B)}')
```
