1) Faça um programa que peça um número em centímetros e o transforme em metros

```python
centimetro = int(input("Digite um número em centímetros: "))

metros = centimetro / 100
print(f"Este valor em metros é {metros}")
```

2) Faça um programa que peça a temperatura em graus Farenheit, transforme e mostre a temperatura em graus Celsius

```python
temperatura = int(input(f"Digite o valor da temperatura em Farenheit: "))

celsius = (temperatura - 32) / 1,8
print(f"A temperatura em graus Celsius é {celsius}")
```

3) Faça um programa que leia 5 números e informe a soma e a média dos números

```python
def calcular_soma_media():
    numeros = []

    for i in range(5):
        while True:
            try:
                numero = float(input(f"Digite o {i + 1}º número: "))
                numeros.append(numero)
                break
            except ValueError:
                print("Inválido. Por favor, insira um número válido.")

    soma = sum(numeros)
    media = soma / len(numeros)

    print(f"A soma dos números é: {soma:.2f}")
    print(f"A média dos números é: {media:.2f}")

calcular_soma_media()
```

4) Faça um programa que peça dois números e imprima o maior deles

```python
def maior_numero():

    while True:
        try:
            numero1 = float(input("Digite o primeiro número: "))
            break
        except ValueError:
            print("Inválido. Por favor, insira um número válido.")

    while True:
        try:
            numero2 = float(input("Digite o segundo número: "))
            break
        except ValueError:
            print("Entrada inválida. Por favor, insira um número válido.")

    if numero1 > numero2:
        print(f"O maior número é: {numero1}")
    elif numero2 > numero1:
        print(f"O maior número é: {numero2}")
    else:
        print("Os números são iguais.")

maior_numero()
```

5) Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido

```python
def pedir_nota():
    while True:
        try:
            nota = float(input("Digite uma nota entre 0 e 10: "))
            if 0 <= nota <= 10:
                print(f"Nota válida: {nota}")
                break
            else:
                print("Nota inválida. A nota deve estar entre 0 e 10.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número.")

pedir_nota()
```

7) Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por um.

```python
def eh_primo(numero):
    if numero <= 1:
        return False
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False
    return True

numero = int(input("Digite um número inteiro: "))

if eh_primo(numero):
    print(f"{numero} é um número primo.")
else:
    print(f"{numero} não é um número primo.")
```
