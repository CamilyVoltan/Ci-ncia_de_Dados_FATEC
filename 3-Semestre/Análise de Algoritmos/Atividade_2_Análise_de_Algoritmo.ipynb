**Bichinho Virtual**

```python
def bichinho(f, s):
  if f <= 2:
    print('Seu bichinho está de mau humor, pois está com fome!')
  elif f == 3:
    print('Fique atento seu bichinho está quase com fome.')
  else:
    print('Seu bichinho está muitoo feliz, pois você alimentou ele muito bem!!')
  if s <= 2:
    print('Seu bichinho está de mau humor, pois pode morrer!')
  elif s == 3:
    print('Fique atento seu bichinho está em uma linha tenue entra a vida e a morte!!!')
  else:
    print('Seu bichinho está muitoo feliz, pois você alimentou ele muito bem e também está saudavel.')




nome = input('Digite um nome para o seu Bichinho Virtual: ')
fome = int(input('Digite o nível de fome de 1 a 5: '))
saude = int(input('O nível de saúde de 1 a 5: '))
idade = int(input('Digite a idade do seu bichinho '))

bichinho(fome, saude)
```

**Busca Binária**

```python
def adivinhar_numero(n):
    baixo = 1
    alto = n
    tentativas = 0

    print(f" Pense em um número entre {baixo} e {alto}. ")

    while baixo <= alto:
        tentativas += 1
        meio = (baixo + alto) // 2
        resposta = input(f"O número é maior que {meio}? (s/n): ").strip().lower()

        if resposta == 's':
            baixo = meio + 1
        elif resposta == 'n':
            alto = meio
        else:
            print("Por favor, responda com 's' para sim ou 'n' para não.")
            tentativas -= 1  # Não conta a tentativa inválida

        if baixo == alto:
            print(f"Seu número é {baixo}!")
            break

    print(f"Descobri seu número em {tentativas} tentativas.")

# Exemplo de uso
n = int(input("Digite o valor máximo n: "))
adivinhar_numero(n)
```
