```python
mport mysql.connector

# Abrimos uma conexão com o banco de dados:
conexao = mysql.connector.connect(host='localhost', database='cadastro', user='root', password='')

if conexao.is_connected():
    db_info = conexao.get_server_info()
    print("Conectado ao servidor Mysql versão ", db_info)

# Cria um cursor para fazer interação:
    cursor = conexao.cursor()
    nome = input("Digite o nome: ")
    dataCompra = input("Digite a data da compra (YYYY-MM-DD): ")
    valorPago = float(input("Digite o valor pago: "))
    localCompra = input("Digite o local da compra: ")

insert_query = """INSERT INTO clientes (nome, dataCompra, valorPago, localCompra)
                        VALUES (%s, %s, %s, %s)"""

# Executa o comando:
cursor.execute(insert_query, (nome, dataCompra, valorPago, localCompra))
# conexao.comit() # quando colocado o 'comit' que executa a operação

print("Dados inseridos com sucesso!")
```
