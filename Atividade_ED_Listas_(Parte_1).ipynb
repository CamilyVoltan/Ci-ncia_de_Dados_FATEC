1. Escreva um programa que recebe como entrada um número inteiro positivo n. Em seguida, seu programa deve ler n números inteiros e adicioná-los em uma lista. Por fim, seu programa receberá um número inteiro x e deve verificar se x pertence ou não à lista.

```python
n = int(input("Digite o número de elementos: "))
lista = []
for i in range(n):
  elemento = int(input("Digite o elemento: "))
  lista.append(elemento)
x = int(input("Digite o elemento a ser verificado: "))
if x in lista:
  print("O elemento pertence à lista.")
else:
  print("O elemento não pertence à lista.")
```

2. Escreva um programa que leia números positivos e os armazene numa lista (até que um número não positivo seja fornecido). Por fim, seu programa receberá um número inteiro x e deve verificar se x pertence ou não à lista.


```python
lista = []
while True:
  num = int(input('Digite um número: '))
  if num <= 0:
    break
  lista.append(num)

x = int(input('Digite um número para verificar: '))
if x in lista:
  print('O número pertence à lista.')
else:
  print('O número não pertence à lista.')
```
