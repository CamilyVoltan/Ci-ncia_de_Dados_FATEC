```python
import matplotlib.pyplot as plt

jogador = ['Amoroso', 'Bebeto', 'Dodô', 'Denilson', 'Ronaldo']
gols = [3, 3, 4, 5, 2]

plt.bar(jogador, gols, color = 'red')
plt.xlabel('JOGADORES')
plt.ylabel('QUANTIDADE DE GOLS')
plt.title('ARTILHEIROS DA RODADA', color = 'blue')
plt.grid(False)
plt.show()
```

GRÁFICO 2

```python
import matplotlib.pyplot as plt
import numpy as np

homens = [9, 6, 9, 8, 5]
mulheres = [6, 8, 9, 6, 7]
# Cria eixo x para homens e mulheres com uma separação de 0.25

# entre as barras
x1 = np.arange(len(homens))
x2 = [x + 0.25 for x in x1]
plt.style.use('bmh') # define o estilo

plt.xlabel('DIAS DA SEMANA ANALISADOS')
plt.ylabel('TOTAL DE ALUNOS NO DIA')
# Plota as barras
plt.bar(x1, homens, width=0.25, label='Homens', color='blue')
plt.bar(x2, mulheres, width=0.25, label='Mulheres', color='pink')

# coloca o nome dos meses como label do eixo x
dias = ['SEG', 'TER', 'QUA', 'QUI', 'SEX']
plt.xticks([x + 0.25 for x in range(len(homens))], dias)

# inseri uma legenda no gráfico
plt.legend()
plt.title("FREQUÊNCIA DOS ALUNOS NA SEMANA")
plt.show()
```
