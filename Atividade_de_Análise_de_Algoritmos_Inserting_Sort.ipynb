### Estrutura do Código

1. **Dicionário `valores_truco`**:
   - Esse dicionário armazena os valores das cartas do truco, onde cada carta (como '1', '2', '3', etc.) é uma chave e seu valor é a prioridade para ordenação.
   - **Exemplo**:
     - '1' (Zap) é a carta mais alta, então recebe o valor 14.
     - '2' (Duque) recebe o valor 13, e assim por diante.

2. **Função `insertion_sort(cartas)`**:
   - Recebe uma lista de cartas como argumento e as ordena in-place (ou seja, sem criar uma nova lista).

### Passo a Passo do Algoritmo

#### 1. Início da Função
- **`for i in range(1, len(cartas))`**:
  - O loop começa no segundo elemento (índice 1) da lista de cartas, pois o primeiro elemento é considerado já ordenado.

#### 2. Definindo a Chave
- **`chave = cartas[i]`**:
  - A carta na posição `i` é armazenada como "chave". Essa carta será inserida na posição correta na parte ordenada da lista.

#### 3. Comparação e Movimentação
- **`while j >= 0 and valores_truco[cartas[j]] > valores_truco[chave]`**:
  - Um segundo loop é iniciado para comparar a "chave" com as cartas anteriores (da esquerda para a direita).
  - `j` começa como `i - 1` (o índice da carta anterior).
  - Se a carta anterior (em `cartas[j]`) for maior que a "chave" de acordo com o dicionário, movemos essa carta uma posição à frente.

#### 4. Inserindo a Chave
- **`cartas[j + 1] = chave`**:
  - Quando encontramos a posição correta (ou `j` se torna menor que 0), inserimos a "chave" na posição correta.

#### 5. Repetição
- O loop externo continua até que todas as cartas sejam processadas.

### Finalização
- Após o loop, a lista de cartas estará ordenada.

```python
# Definindo a ordem das cartas no truco
valores_truco = {        # Dicionário para definir o número da string
    '1': 14,  # Zap
    '2': 13,  # Duque
    '3': 12,  # Terno
    '4': 4,   # Quatro
    '5': 5,   # Cinco
    '6': 6,   # Seis
    '7': 7,   # Sete
    'J': 8,   # Valete
    'Q': 9,   # Dama
    'K': 10   # Rei
}

def insertion_sort(cartas):      # Aqui recebe uma lista de cartas como argumento e as ordena in-place (ou seja, sem criar uma nova lista)
    # Inicia a partir do segundo elemento
    for i in range(1, len(cartas)):
        chave = cartas[i]  # A carta que será inserida
        j = i - 1

        # Move as cartas que são maiores que a chave para uma posição à frente
        while j >= 0 and valores_truco[cartas[j]] > valores_truco[chave]:
            cartas[j + 1] = cartas[j]
            j -= 1

        # Insere a chave na posição correta
        cartas[j + 1] = chave

# Exemplo de uso
cartas_recebidas = ['5', '3', '2', '1', '4']  # Usando valores como strings
print("Cartas antes da ordenação:", cartas_recebidas)

insertion_sort(cartas_recebidas)
print("Cartas após a ordenação:", cartas_recebidas)
```
