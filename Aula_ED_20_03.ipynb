As tuplas, uma vez criadas, não permitem modificações.
- Ou seja, tuplas são **listas imutáveis**;
- Podemos declarar uma tupla utilizando ();


```python
variavel = (elemento_1, elemento_2, ..., elemento_n)
```

- Também podemos declarar uma tupla de maneira explícita
utilizando a função tuple;

```python
variavel = tuple([elemento_1, elemento_2, ..., elemento_n])
```

Declaração implícita:

```python
t1 = ("a")
t2 = ("a",)
print(type(t1), type(t2))
# <class 'str'> <class 'tuple'>
t3 = ("a", "b", "c")
print(t3)
# ('a', 'b', 'c')
```

Declaração explícita:

```python
t1 = tuple([2019])
print(t1)
# (2019,)
t2 = tuple("MC102")
print(t2)
# ('M', 'C', '1', '0', '2')
# Strings também podem ser tuplas de caracteres
```
