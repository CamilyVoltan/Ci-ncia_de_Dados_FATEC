2º Desafio: Análise de Eficiência do Motor

​

A eficiência do veículo depende de muitas variáveis. inclusive fatores externos como topografia, carga e tipo de operação. Mas não consideramos estas variáveis externas e analisamos apenas os dados de operação do veículo.

Quando o ponteiro do conta-giros está na área verde, significa que o veículo está se movimentando com a marcha certa, com o esforço correto e consumo de diesel ideal. Mas dentro ou mesmo fora da área verde, ele ainda pode ter maior economia.

​

Proposta:

Detectar parâmetros e faixas/intervalos que geram maior economia de combustível.

​​

Para o desafio foram selecionados parâmetros que já são conhecidos pelo motorista e ele consegue controlar pela forma de conduzir.
Esse parâmetros são:

Rotação/RPM - Visualiza pelo Conta-Giros no painel do caminhão;

 Posição do pedal do Acelerador - De 0% a 100%.

 Freio motor - 0-Inativo e 1-Ativo - Freio que segura na tração, inverso da força do motor. Auxilia às frenagens longas em trechos onde há declives.

 Inércia - Quando o motorista tira o pé do acelerador, mas o veículo continuará andando engrenado e sem injeção de combustível.
