```python
import pulp as pl

# ------------------ 1 ------------------
prob1 = pl.LpProblem("Velas", pl.LpMaximize)
L = pl.LpVariable("Lavanda", lowBound=0, cat="Integer")
B = pl.LpVariable("Baunilha", lowBound=0, cat="Integer")
C = pl.LpVariable("Canela", lowBound=0, cat="Integer")
prob1 += 6*L + 3*B + 4*C
prob1 += 5*L + 4*B + 6*C <= 180
prob1 += 6*L + 3*B + 5*C <= 210
prob1.solve()
print("Ex1:", {v.name: v.varValue for v in prob1.variables()}, "Lucro=", pl.value(prob1.objective))
```

```python
prob2 = pl.LpProblem("Softwares", pl.LpMaximize)
S1 = pl.LpVariable("S1", lowBound=0, cat="Integer")
S2 = pl.LpVariable("S2", lowBound=0, cat="Integer")
prob2 += 12000*S1 + 18000*S2
prob2 += 50*S1 + 70*S2 <= 3600
prob2 += S1 <= 60
prob2 += S2 <= 40
prob2.solve()
print("Ex2:", {v.name: v.varValue for v in prob2.variables()}, "Lucro=", pl.value(prob2.objective))
```

```python
prob3 = pl.LpProblem("Brinquedos", pl.LpMaximize)
D = pl.LpVariable("Bonecas", lowBound=0, cat="Integer")
T = pl.LpVariable("Carrinhos", lowBound=0, cat="Integer")
lucro_D = 25 - (12+10)
lucro_T = 20 - (8+9)
prob3 += lucro_D*D + lucro_T*T
prob3 += 3*D + 1*T <= 90
prob3 += 1*D + 2*T <= 100
prob3 += D <= 50
prob3.solve()
print("Ex3:", {v.name: v.varValue for v in prob3.variables()}, "Lucro=", pl.value(prob3.objective))
```

```python
prob4 = pl.LpProblem("Transporte", pl.LpMinimize)
F = ["F1","F2","F3"]
Lats = ["L1","L2","L3","L4"]
dist = {
    ("F1","L1"):25,("F1","L2"):20,("F1","L3"):18,("F1","L4"):15,
    ("F2","L1"):30,("F2","L2"):22,("F2","L3"):26,("F2","L4"):19,
    ("F3","L1"):15,("F3","L2"):17,("F3","L3"):20,("F3","L4"):12,
}
supply = {"F1":60,"F2":70,"F3":100}
demand = {"L1":40,"L2":60,"L3":50,"L4":80}
x = pl.LpVariable.dicts("x",(F,Lats),lowBound=0,cat="Continuous")
prob4 += pl.lpSum(dist[i,j]*x[i][j] for i in F for j in Lats)
for i in F: prob4 += pl.lpSum(x[i][j] for j in Lats) <= supply[i]
for j in Lats: prob4 += pl.lpSum(x[i][j] for i in F) == demand[j]
prob4.solve()
print("Ex4:", {(i,j): x[i][j].varValue for i in F for j in Lats}, "Distância=", pl.value(prob4.objective))
```

```python
prob5 = pl.LpProblem("Racoes", pl.LpMaximize)
Rex = pl.LpVariable("Rex", lowBound=0, cat="Integer")
Miau = pl.LpVariable("Miau", lowBound=0, cat="Integer")
prob5 += 25*Rex + 35*Miau
prob5 += 4*Rex + 3*Miau <= 25000
prob5 += 2*Rex + 5*Miau <= 15000
prob5.solve()
print("Ex5:", {v.name: v.varValue for v in prob5.variables()}, "Lucro=", pl.value(prob5.objective))
```
