IDEIA DE CÓDIGO PARA O PROJETO - GERENCIADOR DE TAREFAS DIÁRIAS PESSOAIS

TRABALHO APRESENTADO NA AULA
GERENCIADOR DE TAREFA


```python
# Declara uma lista para adicionar uma tarefa
lista = []

# Pede ao usuário para adicionar a uma tarefa
while True:
    print("Por favor, informe a sua tarefa: ")
    tarefa = input()

    # Pede ao usuário para informar o nível de prioridade da tarefa
    print("Por favor, informe qual o nível de prioridade, se é urgente, médio prazo ou longo prazo: ")
    prioridade = input()

    # Pede ao usuário para informar o status da tarefa
    print("Por favor, informe se o status é:  a fazer, em andamento ou concluído")
    status = input()
    if prioridade == "urgente":
        print("Faça a sua tarefa o quanto antes!")
    if prioridade == "médio prazo":
        print("Você pode concluir em alguns dias!")
    if prioridade == "longo prazo":
        print("Ainda há tempo para planejar!")

    # Cria um dicionário para armazenar as informações da tarefa
    tarefa = {
        "tarefa": tarefa,
        "prioridade": prioridade,
        "status": status
    }
    # Adiciona a tarefa à lista de tarefas
    lista.append(tarefa)

    # Pede ao usuário se deseja adicionar outra tarefa
    print("Deseja adicionar outra tarefa? (s/n)")
    resposta = input()

    if resposta != "s":
        break

# Imprime a lista de tarefas
for tarefa in lista:
    print(tarefa)
```
